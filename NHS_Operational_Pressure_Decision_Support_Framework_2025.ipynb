```python
import pandas as pd
import numpy as np

def month_kpis(csv_path: str):
    df = pd.read_csv(csv_path)

    # Keep only Incomplete Pathways (waiting list snapshot)
    inc = df[df["RTT Part Description"] == "Incomplete Pathways"].copy()

    # waiting-time bin columns
    wait_cols = [c for c in inc.columns if c.startswith("Gt ")]

    # make numeric safely
    for c in wait_cols:
        inc[c] = (
            inc[c].astype(str)
            .str.replace(",", "", regex=False)
            .str.strip()
        )
        inc[c] = pd.to_numeric(inc[c], errors="coerce")

    national_bins = inc[wait_cols].sum(numeric_only=True)

    # build distribution table
    waiting_df = pd.DataFrame({"Bin": national_bins.index, "Patients": national_bins.values})

    # extract week start (handles both Gt 00 To 01 and Gt 104 Weeks)
    weeks = waiting_df["Bin"].str.extract(r"Gt\s+(\d+)\s+To", expand=False)
    waiting_df["Weeks"] = pd.to_numeric(weeks, errors="coerce")
    waiting_df = waiting_df.dropna(subset=["Weeks"]).copy()
    waiting_df["Weeks"] = waiting_df["Weeks"].astype(int)
    waiting_df = waiting_df.sort_values("Weeks").reset_index(drop=True)

    total = float(waiting_df["Patients"].sum())
    over_18 = float(waiting_df.loc[waiting_df["Weeks"] > 18, "Patients"].sum())
    over_52 = float(waiting_df.loc[waiting_df["Weeks"] > 52, "Patients"].sum())

    pct_18 = (over_18 / total) * 100 if total > 0 else np.nan
    pct_52 = (over_52 / total) * 100 if total > 0 else np.nan

    return total, pct_18, pct_52
```

```python
import pandas as pd
import numpy as np
import re

def specialty_hotspots(csv_path, month_label):
    df = pd.read_csv(csv_path)

    # Focus on backlog snapshot
    df_inc = df[df["RTT Part Description"] == "Incomplete Pathways"].copy()

    # waiting-time bin columns
    wait_cols = [c for c in df_inc.columns if c.startswith("Gt ")]

    # Clean numeric
    for c in wait_cols:
        df_inc[c] = (df_inc[c].astype(str).str.replace(",", "", regex=False).str.strip())
        df_inc[c] = pd.to_numeric(df_inc[c], errors="coerce").fillna(0)

    # Map bins to starting week number
    def start_week(colname):
        m = re.search(r"Gt\s+(\d+)\s+To", colname)
        if m:
            return int(m.group(1))
        m2 = re.search(r"Gt\s+(\d+)\s+Weeks", colname)
        return int(m2.group(1)) if m2 else None

    week_map = {c: start_week(c) for c in wait_cols}
    wait_cols = [c for c in wait_cols if week_map[c] is not None]

    # Group by specialty
    group_col = "Treatment Function Name"
    g = df_inc.groupby(group_col)[wait_cols].sum()

    total = g.sum(axis=1)

    over18_cols = [c for c in wait_cols if week_map[c] > 18]
    over52_cols = [c for c in wait_cols if week_map[c] > 52]

    over18 = g[over18_cols].sum(axis=1) if over18_cols else 0
    over52 = g[over52_cols].sum(axis=1) if over52_cols else 0

    out = pd.DataFrame({
        "Month": month_label,
        "Specialty": total.index,
        "Total_Waiting": total.values,
        "Over_18": np.array(over18),
        "Over_52": np.array(over52),
    })

    out["Pct_Over_18"] = np.where(out["Total_Waiting"] > 0, out["Over_18"] / out["Total_Waiting"] * 100, 0)
    out["Pct_Over_52"] = np.where(out["Total_Waiting"] > 0, out["Over_52"] / out["Total_Waiting"] * 100, 0)

    # remove tiny specialties (helps signal)
    out = out[out["Total_Waiting"] >= 1000].copy()

    return out

all_specialty = []
for month, path in month_files.items():
    all_specialty.append(specialty_hotspots(path, month))

specialty_df = pd.concat(all_specialty, ignore_index=True)

specialty_df.shape, specialty_df.head()
```

# NHS RTT Operational Pressure Framework (Apr–Dec 2025)

## Project Overview
This project builds a decision-support analytics prototype using NHS England RTT (Referral to Treatment) waiting times data.  
Using monthly national extracts from April to December 2025, the analysis tracks backlog pressure and long-wait breaches, and proposes an operational escalation model to support capacity planning and governance-led decision-making.

**Scope:** Non-clinical, aggregate reporting only (no patient-level data).  
**Output:** A structured monitoring framework + recommendations for NHS operational pressure management.


## Executive Summary (Apr–Dec 2025)

This project converts NHS RTT waiting list data into an operational pressure monitoring prototype.

## Operational Metrics Defined

### KPI 1 — Total Waiting
Total number of patients on incomplete RTT pathways nationally (all waiting-time bins summed).

### KPI 2 — % Over 18 Weeks
Share of patients waiting longer than 18 weeks:
- Indicator of breach pressure against RTT standards.

### KPI 3 — % Over 52 Weeks
Share of patients waiting longer than 52 weeks:
- Indicator of extreme backlog / harm-risk cohort.

### Composite — Pressure Score (0–1)
A weighted composite measure designed for decision-support monitoring.

Example weighting:
- 0.5 × Normalised Total Waiting
- 0.3 × Normalised % Over 18 Weeks
- 0.2 × Normalised % Over 52 Weeks

The score is not a national official metric; it is a **prototype** to support consistent monitoring and escalation logic.

## Operational Escalation Model (Prototype)
### Escalation Tiers Based on Pressure Score

To operationalise the framework, a tiered escalation model is proposed:

Tier 1 – Stable Pressure  
Pressure Score < 0.40  
Routine monitoring. No immediate intervention required.

Tier 2 – Moderate Pressure  
Pressure Score 0.40 – 0.70  
Operational review recommended.  
Specialty-level capacity adjustments considered.

Tier 3 – High Pressure  
Pressure Score > 0.70  
Escalation to regional management level.  
Short-term surge planning or elective reprioritisation may be required.
